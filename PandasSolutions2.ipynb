**Question 2)** In the column sex, Replace male/female with the 2 values 1/0<br><br>

**Question 5)** ** The embarked attribute is having missing values, replace them by the most occurred value in this column.<br><br>


```python
# Embarked attribute is having missing values, best way to fill this would 
# be by most occurred value in this column

data['embarked'].unique()
data['embarked'].value_counts()
nullindexes=data['embarked'].isnull()
data.loc[nullindexes, "embarked"]
# data["embarked"].mode() # Always returns Series even if only one value is returned.
data["embarked"]=data["embarked"].fillna(data["embarked"].mode()[0])
data.loc[nullindexes, "embarked"]
data['embarked'].unique()
```

**Question 6)** ** Fare is a numeric attribute and it has missing values too. <br><br>
	In case of Fare, filling missing values with median will be better choice than mean or mode.<br><br>

```python
# Fare is a numeric attribute and it has missing values too. 
# In case of Fare, filling missing values with median will be better choice than mean or mode
# Create a new column FareCategory to categorized fare into 4 categories: 0,1,2,3 (use pd.qcut())

nullindexes=data['fare'].isnull()
data["fare"]=data["fare"].fillna(data["fare"].median())
data.loc[nullindexes, "fare"]

data['fare'].value_counts()

# print(data.columns)
# #print(pd.qcut(data["fare"], 4).unique())

```

**Question 7)** ** Create a new column FareCategory to categorized fare into 4 categories: 0,1,2,3 (use pd.qcut())<br><br>



```python

data["farecategory"]=pd.qcut(data["fare"], 4, labels=False)
data["farecategory"].unique()
data["farecategory"].head(15)
```


**Question 8)** Print the survival rate per fare category<br><br>

```python
# print the survival rate per fare category:
# print(data.columns)

data[["farecategory", "survived"]].groupby("farecategory").mean()
```

**Question 9)** ** In the Titanic dataset, many of the Age values were given as NaN. <br><br>
   Replace the NaN value with the median age for each gender within each passenger class.<br><br>


```python
#In the Titanic dataset, many of the Age values were given as NaN. 
#Replace the NaN value with the median age for each gender within each passenger class.

# Create new column AgeFill to put values into. 
data['ageFill'] = data['age']

# This retains the state of the original data.
indexes=data['age'].isnull()

data['ageFill'].fillna(data.groupby(["sex", "pclass"])["age"].transform("median"), inplace=True)

# To print the 10 first rows with age missing
data[indexes].filter(["ageFill", "age", "sex", "pclass"]).head(10)
```


**Question 10)** Create a new column that records whether the Age was originally missing<br><br>


```python
# Create a feature that records whether the Age was originally missing
data['ageIsNull'] = pd.isnull(data['age']).astype(int)
data['ageIsNull'].head(20)

data.head(20)
```
